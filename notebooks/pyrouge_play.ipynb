```python
def createSummaries(sum_algo, abs_path, out_path, multiDocument = False, k = 5):
    # Extracted from the input folder name 
    setID = os.path.dirname(abs_path)
    
    # Stores DocumentIDs
    docIDs = []
    
    # Create document set that we wish to evaluate
    D = []
    for filename in open(abs_path):
        # Only take files that we've parsed before!
        tmp = filename.split('.')
        if tmp[0] == 'parsed':
            docIDs.append(temp[1])
            filepath = os.join.path(abs_path, filename)
            with open(filepath) as inputDoc:
                D.append(inputDoc.readlines())
            
    # Pass this to the algorithm which should return the summary as
    # a list of sentences.
    if multiDocument:
        summary = sum_algo(D, k, multiDocument)
        # Write out the summary
        filepath = os.join.path(out_path, "Summary.{}.txt".format(setID))
        with open(filepath) as out:
            for s in summary:
                out.write("{}.\n".format(s))
    else:
        for i in range(len(D)):
            summary = sum_algo(D[i], k, multiDocument)
            filepath = os.join.path(out_path, "Summary.{}.txt".format(docIDs[i]))
            with open(filepath) as out:
                for s in summary:
                    out.write("{}.\n".format(s))
```
